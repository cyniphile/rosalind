```python
def find_subs(string, substr):
    indicies = []
    for i, char in enumerate(string):
        test = string[i: i+len(substr)]
        if test == substr:
            indicies.append(str(i+1))
    return indicies
```

```python
with open('./input.txt') as f:
    lines = f.readlines()
with open('./answer.txt', 'w') as a:
    for i in range(1, len(lines), 2):
        string = lines[i]
        substr = lines[i+1].strip()
        answer = find_subs(string, substr)
        a.writelines(" ".join(answer)+"\n")
        print(i)
        print(string)
        print(substr)
        print(answer)
```
